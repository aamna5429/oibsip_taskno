```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv('covid 19.csv')
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns = df.columns.str.lower()
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y="Region",data=df)
plt.show()
```

```python
df['estimated unemployment rate (%)']
```

```python
avg_unemployment_rate=df.groupby('region')['estimated unemployment rate (%)'].mean()
state_with_highest_unemployment=avg_unemployment_rate.idxmax()
high_unemployment_rate=avg_unemployment_rate.max()
state_with_lowest_unemployment=avg_unemployment_rate.idxmin()
low_unemploy_rate=avg_unemployment_rate.min()
print(f"state  with  high employment : {state_with_highest_unemployment}")
print(f"high employment rate : {high_unemployment_rate}")
print(f"state  with  low employment : {state_with_lowest_unemployment}")
print(f"low employment rate : {low_unemploy_rate}")
```

```python
regions = avg_unemployment_rate.index
avg_rates = avg_unemployment_rate.values

plt.figure(figsize=(10, 6))
plt.bar(regions, avg_rates, color='skyblue')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate')
plt.title('Average Unemployment Rate by Region')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='estimated unemployment rate (%)', data=df,palette="Set1")
plt.title("Unemployment Rate Distribution by Region")
plt.xlabel("State")
plt.ylabel("Estimated Unemployment Rate")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='estimated labour participation rate (%)', data=df,palette="hsv")
plt.title("Labour_Participation_ Rate")
plt.xlabel("Region")
plt.ylabel("Labour_Participation_ Rate")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='estimated employed', data=df,palette="brg")
plt.title("estimated_employee")
plt.xlabel("Region")
plt.ylabel("estimated_employee")
plt.xticks(rotation=90)
plt.show()
```

```python
pair=df[["estimated unemployment rate (%)","estimated employed","estimated labour participation rate (%)"]]
sns.pairplot(pair,markers="*",palette="winter")
plt.show()
```

```python
df["area"].value_counts()
```

```python
# Group the data by Region and Area, and calculate the mean unemployment rate
mean_unemployment = df.groupby(['region', 'area'])['estimated unemployment rate (%)'].mean().unstack()

# Plotting the graph
plt.figure(figsize=(12,6))
plt.plot(mean_unemployment.index, mean_unemployment['Urban'], marker='o', label='Urban', color='blue')
plt.plot(mean_unemployment.index, mean_unemployment['Rural'], marker='o', label='Rural', color='orange')

# Add labels and title
plt.title('Comparison of Unemployment Rate Mean between Urban and Rural')
plt.xlabel('Regions')
plt.ylabel('Unemployment Rate Mean')
plt.xticks(rotation=90)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df.columns = df.columns.str.strip()

# Strip leading/trailing spaces in the 'Date' column
df['date'] = df['date'].str.strip()

# Convert 'Date' column to datetime format
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y', errors='coerce')

# Group by 'Date' and calculate the mean unemployment rate across regions
unemployment_over_time = df.groupby('date')['estimated unemployment rate (%)'].mean()

# Plot the bar chart
plt.figure(figsize=(12,6))
unemployment_over_time.plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('Change in Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Average Unemployment Rate (%)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
